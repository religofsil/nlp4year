```python
import nltk
from nltk.collocations import *
bigram_measures = nltk.collocations.BigramAssocMeasures()
stopwords=nltk.corpus.stopwords.words('russian')
```

```python
from string import punctuation
with open('war.txt') as f:
    text = f.read().lower()
    translation = str.maketrans("","", punctuation)
    text = text.lower().translate(translation)
```

```python
words = text.split()
```

```python
import pymorphy2
morph = pymorphy2.MorphAnalyzer()
words = [morph.parse(word)[0].normal_form for word in words]
```

```python
finder = BigramCollocationFinder.from_words(words)
finder.apply_word_filter(lambda w: len(w) < 3 or w.lower() in stopwords)
```

```python
from nltk.metrics.spearman import *
```

```python
scored_bigrams = finder.score_ngrams(bigram_measures.student_t)
scored_bigrams=[i for i in scored_bigrams if 'друг' in i[0]]
scored_results_student=ranks_from_scores(sorted(scored_bigrams, key = lambda x: x[1], reverse=True))
sorted(scored_bigrams, key = lambda x: x[1], reverse=True)
```

```python
print(finder.score_ngram(bigram_measures.student_t, 'верный', 'друг'))
print(finder.score_ngram(bigram_measures.student_t, 'друг', 'друг'))
print(finder.score_ngram(bigram_measures.student_t, 'меланхолия', 'друг'))
```

```python
print(finder.score_ngram(bigram_measures.raw_freq, 'верный', 'друг'))
print(finder.score_ngram(bigram_measures.raw_freq, 'друг', 'друг'))
print(finder.score_ngram(bigram_measures.raw_freq, 'меланхолия', 'друг'))
```

Я использовала распределение Стьюдента и "сырую" частотность на первых двух томах "Войны и мира" в качестве корпуса. "Меланхолию друга" nltk ожидаемо не считает коллокацией, "верный друг" неожиданно тоже не встречается в корпусе. "Друг друга" зато встречается очень часто.
